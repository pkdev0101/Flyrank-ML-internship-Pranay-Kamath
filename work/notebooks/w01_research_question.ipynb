## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import sys
import subprocess

REPO_URL = "https://github.com/pkdev0101/Flyrank-ML-internship-Pranay-Kamath"
REPO_DIR = "/content/Flyrank-ML-internship-Pranay-Kamath"

if "google.colab" in sys.modules:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(
            ["git", "clone", "--depth", "1", REPO_URL, REPO_DIR],
            check=True
        )
    os.chdir(REPO_DIR)

import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

declining = df["trend_direction"].str.lower().eq("down")

stale_visible = (
    (df["days_since_last_update"] >= 180)
    & (df["impressions_90d"] >= 500)
)

low_ctr_visible = (
    (df["impressions_90d"] >= 500)
    & (df["avg_position"].between(1, 20))
    & (df["ctr"] < 0.5)
)

print(f"Total pages: {len(df):,}")
print(f"Declining pages: {declining.sum():,} ({declining.mean():.1%})")
print(f"Stale and visible pages: {stale_visible.sum():,}")
print(f"Visible pages with low CTR: {low_ctr_visible.sum():,}")
```

The starter dataset has 30,000 pages, and 16,262 of them, or 54.2%, are labeled as declining. Only 17 pages are both stale and highly visible under my current rule, suggesting that this rule may be too strict by itself. There are 9,745 visible pages with low CTR, showing that there are many possible review candidates. These numbers support using a ranked scoring system so an editor can focus on the strongest opportunities instead of reviewing thousands of pages equally.
